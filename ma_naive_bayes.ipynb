```python
from sklearn.cluster import KMeans
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt 
from sklearn.preprocessing import StandardScaler
import plotly.graph_objects as go
```

### read data

```python
stocks = pd.read_csv("data/SPY_daily.csv")
```

### engineer features

```python
#create performance features
stocks['high_adj'] = (stocks['High'] - stocks['Close']) / stocks['Close']
stocks['low_adj'] = (stocks['Low'] - stocks['Close']) / stocks['Close']
stocks['body'] = (stocks['Close'] - stocks['Open']) / stocks['Open']
stocks['close_prev'] = stocks['Close'].shift(1)

#create MA features
stocks['ma_7'] = stocks['Close'].rolling(window=7).mean()
stocks['ma_3'] = stocks['Close'].rolling(window=3).mean()
#stocks = stocks.iloc[7:]
stocks['change'] = (stocks['Close'] - stocks['close_prev']) / stocks['close_prev']

#stocks['pct_above_ma_7'] = stocks['ma_7'] - stocks['']
#stocks['pct_above_ma_3'] = 
```

```python
long = False

def take_action(s, ma):
    global long
    if s['ma_' + ma] < s['Close']:
        if long: 
            long = False
            return -1
        if not long: return 0
    if s['ma_' + ma] >= s['Close']:
        if long: return 0
        if not long: 
            long = True
            return 1

stocks['action_ma_7'] = stocks.apply(take_action, axis=1, args=('7',))
stocks['action_ma_3'] = stocks.apply(take_action, axis=1, args=('3',))
```

```python
sample = stocks.iloc[10:100]
data = [go.Candlestick(
                x=sample['Date'],
                open=sample['Open'],
                high=sample['High'],
                low=sample['Low'],
                close=sample['Close'],
                name='price'),
                go.Scatter(x=sample['Date'],
                y=sample['ma_7'],
                mode="lines",
                line={'color': 'cyan'},
                name='MA 7')]
annotations = []
fig = go.Figure(data=data)


fig.show()
```
